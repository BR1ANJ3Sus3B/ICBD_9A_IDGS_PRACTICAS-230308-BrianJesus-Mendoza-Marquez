15-. Conteo de pokemons por atributo Tipo (Graficado).
Se usa la librería matplot para configurar el tamaño de la gráfica. NOTA: *Por el número de pokemones de tipo Agua y Normal es alto comparado al resto de los pokemones respectivos tendrán siempre un total de stats más alto

```python
tipos_ordenados = sorted(pokemons1['Type'].unique())

plt.figure(figsize=(10, 5))
ax = sns.countplot(
    data=pokemons1,
    x='Type',
    hue='Type',
    palette='twilight',
    order=tipos_ordenados,      # Ordena el eje X alfabéticamente
    hue_order=tipos_ordenados   # Ordena las leyendas (colores) alfabéticamente
)

# Agregar totales en la parte superior de cada barra
for container in ax.containers:
    ax.bar_label(
        container,
        label_type='edge',
        padding=3
    )

plt.title('Número de pokemones agrupados por Tipo')
plt.xlabel('Tipo')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

16-. Revisar la distribución de estadistícas con boxplot.
Se confiugura el estilo de la gráfica con el método .set_style.
Además, se utiliza como data el DataFrame de pokemones, pero excluytendo las columnas; #, Total, Generation, y Legendary.

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.boxplot(data=pokemons.drop(['#','Total','Generation','Legendary'],axis=1),fliersize=3,palette='seismic')
plt.title('Boxplots para Estadísticas')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Columnas que se excluirán
columnas_a_excluir = ['#', 'Name', 'Type 1', 'Type 2', 'Total', 'Generation', 'Legendary', 'Armor']

# Eliminar únicamente las columnas que existen
datos_filtrados = pokemons1.drop(
    columns=[col for col in columnas_a_excluir if col in pokemons1.columns]
)

# Ordenar las columnas alfabéticamente
datos_ordenados = datos_filtrados[sorted(datos_filtrados.columns)]

# Configurar estilo
sns.set_style('darkgrid')

# Crear gráfica
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=datos_ordenados,
    palette='cubehelix',
    fliersize=3
)

# Título y etiquetas
plt.title('Boxplot of Pokémon Statistics')
plt.xlabel('Statistics')
plt.ylabel('Values')

# Ajustar y mostrar
plt.tight_layout()
plt.show()
```
