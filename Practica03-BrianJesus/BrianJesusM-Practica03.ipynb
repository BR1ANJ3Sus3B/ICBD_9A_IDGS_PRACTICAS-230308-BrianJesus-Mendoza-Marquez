# **25. Cálculo de Estadísticas Promedio por Tipo**
Regresemos a los gráficos de enjambre (swarmplots), pero esta vez para profundizar en un análisis comparativo directo.
Cada gráfico permite identificar qué tipos de Pokémon presentan los valores promedio más altos o más bajos en cada característica, facilitando el análisis exploratorio de los datos.

```python
import pandas as pd

pok_melt = pd.melt(
    pokemons1,
    id_vars=['Name', 'Type 1', 'Legendary'],
    value_vars=['HP', 'Defense', 'Attack', 'Sp. Atk', 'Sp. Def', 'Speed'],
    var_name='Statistic',
    value_name='Value'
)

pok_melt.head()
```

```python
import pandas as pd

pok_melt = pd.melt(
    pokemons1,
    id_vars=['Name', 'Type 1', 'Legendary'],
    value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'],
    var_name='Statistic',
    value_name='Value'
)

# Cambiar el nombre de la columna para que se llame Type
pok_melt = pok_melt.rename(columns={'Type 1': 'Type'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

list_types = sorted(pok_melt['Type'].unique())

plt.figure(figsize=(17,22))

k = 1

for i in list_types:
    if k > 18:
        break

    plt.subplot(6,3,k)
    k += 1

    datos = pok_melt[pok_melt['Type'] == i]

    sns.swarmplot(
        data=datos,
        x='Statistic',
        y='Value',
        hue='Statistic',
        palette='gist_stern',
        legend=False
    )

    plt.title(i)
    plt.xlabel("")
    plt.ylabel("")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
